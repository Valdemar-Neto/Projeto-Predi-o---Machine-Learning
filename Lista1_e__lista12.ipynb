### Escolhendo as classes

```python
x = dataset.loc[:,['totalAcertL01','aprovadoLista01','subm/questL01','timeInSecoundsL01','time/submL01']].values
y = dataset['submeteuL12'].values

print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
print(X_train[0:5])
sc = MinMaxScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
print(X_train[0:5])
```

### Descobrindo os melhores parâmetros para a svm


```python
import numpy as np
gamma_range = np.logspace(-9, 3, 13)
print(gamma_range)
```

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV

dataset = datasets.load_iris()
X = dataset.data
y = dataset.target

parametros = {'C':[0.1,1,10],'gamma':gamma_range,'kernel':['rbf','linear']}

svm_classificador = svm.SVC()

pesquisa = GridSearchCV(svm_classificador, parametros)
pesquisa.fit(X,y)

print("Melhores parâmetros:",pesquisa.best_params_)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=1, gamma=1e-09, C=1)
classifier.fit(X_train, Y_train)
```

```python
y_pred = classifier.predict(X_test)

print(Y_test[0:])
print(y_pred[0:])
```

```python
from sklearn.metrics import accuracy_score

print("Acurácia: ", accuracy_score(Y_test,y_pred)*100, "%"  )
```

```python
from sklearn.metrics import balanced_accuracy_score
bc = balanced_accuracy_score(Y_test, y_pred)
print(bc*100)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
print("Matrix de confusão: ")
print(cm)
```
