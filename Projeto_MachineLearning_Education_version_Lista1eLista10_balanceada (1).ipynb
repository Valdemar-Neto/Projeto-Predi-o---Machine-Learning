### Dados na educação através do repositório de dados da LOP


```python
import numpy as np
import pandas as pd

lista1e10 = pd.read_csv('/content/drive/MyDrive/Projeto_Machine_Learning_Education/DadosLop.csv')
lista1e10.head()

```

```python
lista2 = pd.read_csv('/content/drive/MyDrive/Projeto_Machine_Learning_Education/lista02.csv')
lista2.head()
```

```python
dataset = pd.merge(lista1e10,lista2,how='outer')
dataset.head()
```

```python
display(dataset)
```

```python
pd.set_option('display.max_columns', None)
dataset
data = dataset.dropna()
print(data)
```

### Selecionando na coluna submeteuL01 os valores igual a zero

```python
coluna_selecionada = data.loc[data['submeteuL10'] == 0,:]
coluna_selecionada1 = data.loc[data['submeteuL10'] == 1,'submeteuL10']
print(len(data['submeteuL10']))
quantidade_1 = len(coluna_selecionada1)
print("quantidade de submissões lista igual a  1",quantidade_1)

quantidade = len(coluna_selecionada)
print("quantidade de submissoes da lista 1 igual a 0",quantidade)
print(coluna_selecionada)
```

### Equilibrando a rede



```python
valores_sorteados = coluna_selecionada.sample(n=98*3, replace=True)
print(valores_sorteados)
```

### Concatenando os valores


```python
todos = pd.concat([data, valores_sorteados], ignore_index=True)
print(todos)
print(todos['submeteuL10'])
```

```python
todos_selecionada = todos.loc[todos['submeteuL10'] == 0,:]
todos_selecionada1 = todos.loc[todos['submeteuL10'] == 1,:]
print(len(todos['submeteuL10']))
print(todos_selecionada.shape)
print(todos_selecionada1.shape)
todos.head()
```

```python
x = todos.loc[:,['totalAcertL01','aprovadoLista01','subm/questL01','timeInSecoundsL01','time/submL01']].values
y = todos['submeteuL10'].values

print(x)
print(y)
```

### Normalização


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

### Normalização

```python
from sklearn.preprocessing import MinMaxScaler
print(X_train[0:5])
sc = MinMaxScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
print(X_train[0:5])
```

### Treinamento com os valores da tebela correlacinando os acertos, o tempo de submissao, a quantidade de listas submetidas e

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', random_state=1, gamma=2, C=40)
classifier.fit(X_train, y_train)
```

### Avaliação do modelo de decisão

```python
y_pred = classifier.predict(X_test)

print(y_test[0:])
print(y_pred[0:])
```

```python
from sklearn.metrics import accuracy_score

print("Acurácia: ", accuracy_score(y_test,y_pred)*100, "%"  )
```

### O balanço médio da acurácia


```python
from sklearn.metrics import balanced_accuracy_score
bc = balanced_accuracy_score(y_test, y_pred)
print(bc*100)
```

### Matriz de confusão


```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Matrix de confusão: ")
print(cm)
```

```python
print(cm[0,0])
print(cm[1,0])
```

### Precisão do modelo

```python
from sklearn.metrics import precision_score
pr = precision_score(y_test,y_pred,average='macro')
print("A precisão do modelo é ",pr*100)
```

```python
from sklearn.metrics import recall_score
rec = recall_score(y_test,y_pred)
print(rec)
```
