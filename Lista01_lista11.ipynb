### Juntando Lista 1 e lista 11

### Tratando os dados, para tirar os valores NaN


```python
data = dataset.dropna()
data
```

### Balanceando a rede

```python
coluna_selecionada = data.loc[data['submeteuL11'] == 0,:]
coluna_selecionada1 = data.loc[data['submeteuL11'] == 1,'submeteuL11']
print(len(data['submeteuL11']))
quantidade_1 = len(coluna_selecionada1)
print("quantidade de submisses lista igual a  1",quantidade_1)

quantidade = len(coluna_selecionada)
print("quantidade de submissoes da lista 1 igual a 0",quantidade)
print(coluna_selecionada)
```

```python
valores_sorteados = coluna_selecionada1.sample(n=30, replace=True)
print(valores_sorteados)
```

```python
todos = pd.concat([data, valores_sorteados], ignore_index=True)
print(todos)
print(todos['submeteuL11'])

dados = todos.fillna(0)
dados.head()
```

### Escolhendo as classes para o treinamento

```python
x = dados.loc[:,['totalAcertL01','aprovadoLista01','subm/questL01','timeInSecoundsL01','time/submL01','submeteuL01']].values
y = dados['submeteuL11'].values

print(x)
print(y)
```

### Normalizao

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
print(X_train[0:5])
sc = MinMaxScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
print(X_train[0:5])
```

### Descobrindo os parametros da svm automaticamente

```python
import numpy as np
gamma_range = np.logspace(-9, 3, 13)
print(gamma_range)
```

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV

dados = datasets.load_iris()
X = dados.data
y = dados.target

parametros = {'C':[0.1,1,10],'gamma':gamma_range,'kernel':['rbf','linear']}

svm_classificador = svm.SVC()

pesquisa = GridSearchCV(svm_classificador, parametros)
pesquisa.fit(X,y)

print("Melhores parmetros:",pesquisa.best_params_)
```

### Treinando a rede svm


```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=1, gamma=1e-09, C=10)
classifier.fit(X_train, Y_train)
```

### Avaliao do modelo

```python
y_pred = classifier.predict(X_test)

print(Y_test[0:])
print(y_pred[0:])
```

```python
from sklearn.metrics import accuracy_score

print("Acurcia: ", accuracy_score(Y_test,y_pred)*100, "%"  )
```

```python
from sklearn.metrics import balanced_accuracy_score
bc = balanced_accuracy_score(Y_test, y_pred)
print(bc*100)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
print("Matrix de confuso: ")
print(cm)
```
